```python
#Basic Operations with a Data Frame (Day 3)
import pandas as pd
stats = pd.read_csv('/Users/coreysolitaire/Desktop/Python A-Z Programing Course/Module 5 (Data Frames)/P4-Demographic-Data.csv')
stats.columns=['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers', 'IncomeGroup']
```

```python
result.head()
```

```python
stats.head()
```

```python
#Removing a column:
#1 specifies x axis, default axis is y (0)
stats.drop('MyCalc', 1).head()
```

```python
#Notice it did not actually drop.....
stats.head()
```

```python
#To fully drop use following prcedure:
stats = stats.drop('MyCalc', 1).head()
stats.head()
```

```python
#Filtering a Data Frame (Day #4):
#Filtering is about Rows.....
import pandas as pd
stats = pd.read_csv('/Users/coreysolitaire/Desktop/Python A-Z Programing Course/Module 5 (Data Frames)/P4-Demographic-Data.csv')
stats.columns=['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers', 'IncomeGroup']
stats.head()
```

```python
stats.InternetUsers < 2
```

```python
#Takes a column and turns the value to a Boolean
Filter = stats.InternetUsers < 2
```

```python
Filter [10:15]
```

```python
#Looks up what is in filter, and returns only True values
stats [Filter]
```

```python
#New problem,
stats.BirthRate > 40
```

```python
Filter2 = stats.BirthRate > 40
```

```python
stats [Filter2]
```

```python
#Sweet Shortcut... do not define function first
stats [stats.BirthRate > 40]
```

```python
#Even more interesting... lets combine both filters...
# Results in Error - stats [Filter and Filter2] 
stats [Filter & Filter2] # This one works, need bit-wise & operator
```

```python
#Shortcut (Must use Brackets)
stats [(stats.BirthRate > 40)&(stats.InternetUsers < 2)]
```

```python
#Introduction to Seaborn
import pandas as pd
stats = pd.read_csv('/Users/coreysolitaire/Desktop/Python A-Z Programing Course/Module 5 (Data Frames)/P4-Demographic-Data.csv')
stats.columns=['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers', 'IncomeGroup']
stats.head()
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 10,5
```

```python
stats.head()
```

```python
#Distribution:
vis1 = sns.distplot(stats['InternetUsers'], hist_kws=dict(edgecolor="k", linewidth=2))

#To turn on histograph outlines: hist_kws=dict(edgecolor="k", linewidth=2)
```

```python
#Lets break the datat up in to more bins....
vis1 = sns.distplot(stats['InternetUsers'], bins =40, hist_kws=dict(edgecolor="k", linewidth=2))
```

```python
#Boxplots
vis2 = sns.boxplot(data=stats, x='IncomeGroup', y='BirthRate')
```

```python
#Seaborn Gallery
```

```python
#Can be prefomred calling variables first
vis3 = sns.lmplot(x='InternetUsers', y='BirthRate', data=stats,)
```

```python
#Or can be preformed calling data first, no difference
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate')
```

```python
#Non-linear relationahip so lets get rid of the regression line..
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False)
```

```python
#Birthrate histogram distribution
vis4 = sns.distplot(stats['BirthRate'], bins =40, hist_kws=dict(edgecolor="k", linewidth=2))
```

```python
#Now lets add some color...

vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', \
                  fit_reg=False, hue= 'IncomeGroup')
```

```python
# Or change the size...

vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', \
                  fit_reg=False, hue= 'IncomeGroup', height=10)
```

```python
#or the aspect ratio....

vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', \
                  fit_reg=False, hue= 'IncomeGroup', height=10, aspect=2)
```

```python
#Keyword arguments in Python (Advanced Tutorial)
```

```python
# How do we change the marker size inside of a visualization?
# Have to access the 's' dictionary and choose a size

vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', \
                  fit_reg=False, hue= 'IncomeGroup', height=10,  \
                  aspect=1, scatter_kws={'s':200})
```
